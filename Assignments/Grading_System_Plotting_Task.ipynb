```python
import pandas as pd
```

```python
import matplotlib.pyplot as plt
```

```python
students_data = pd.read_csv("studentsData.csv")
```

```python
students_data.shape
```

```python
students_data.head()
```

```python
students_data["Total Marks"] = 100
```

```python
students_data["Obtained Marks"] = students_data["exam1"]+students_data["exam2"]+students_data["final_exam"]
```

```python
students_data["Percentage"] = (students_data["Obtained Marks"]/students_data["Total Marks"]) * 100
```

```python
students_data.head()
```

```python
def grades(score):
    if score >= 80:
        return "A"
    elif score >= 70:
        return "B"
    elif score >= 60:
        return "C"
    elif score >= 50:
        return "D"
    else:
        return "F"

students_data["Grades"] = students_data["Percentage"].apply(grades)
```

```python
students_data["Class"] = students_data.roll_no.str.split("/").str[1]
students_data
```

```python
students_group = students_data.groupby("Class")
for student_class in students_data["Class"].unique():
    chart_value = students_group.get_group(student_class).Grades.value_counts()
    plt.title(f"Class: {student_class}")
    plt.xlabel("Grades")
    plt.ylabel("No.of Students")
    plt.bar(chart_value.index,chart_value.values,color=["blue","skyblue","red","purple","yellow"])
    plt.show()
```

```python
# Creating pivot_table of complete student data (summarized-data)
pivot_table = students_data.pivot_table(index="Grades",columns="Class",aggfunc = "size")  
# pivot_table = students_data.pivot_table(index="Grades",columns="Class",aggfunc = "sum")  
```

```python
pivot_table.plot(kind = "bar",colormap= "tab10",title="grades")
plt.legend(title="SMIT")
plt.show()
```
