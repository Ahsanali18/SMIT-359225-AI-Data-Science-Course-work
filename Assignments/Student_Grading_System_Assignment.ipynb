```python
import pandas as pd
```

```python
students_data = pd.read_csv("studentsData.csv")
```

```python
students_data.shape
```

```python
students_data.head()
```

```python
students_data["Total Marks"] = 100
```

```python
students_data["Obtained Marks"] = students_data["exam1"]+students_data["exam2"]+students_data["final_exam"]
```

```python
students_data["Percentage"] = (students_data["Obtained Marks"]/students_data["Total Marks"]) * 100
```

```python
students_data.head()
```

```python
# add a column of class
# students_data["roll_no"]str.split("/").str[1]
```

```python
def grades(score):
    if score >= 80:
        return "A"
    elif score >= 70:
        return "B"
    elif score >= 60:
        return "C"
    elif score >= 50:
        return "D"
    else:
        return "F"

students_data["Grades"] = students_data["Percentage"].apply(grades)
```

```python
students_data["Class"] = students_data.roll_no.str.split("/").str[1]
students_data
```
