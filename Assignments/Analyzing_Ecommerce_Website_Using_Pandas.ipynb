```python
import pandas as pd
```

```python
customers = pd.read_csv("customers.csv")
customers
```

```python
products = pd.read_csv("products.csv")
products
```

```python
orders = pd.read_csv("orders.csv")
orders
```

```python
merged_data = pd.merge(orders, products, on ="product_id")
merged_data = pd.merge(merged_data,customers, on = "customer_id")
print(merged_data)
```

```python
merged_data["total_sale"] = merged_data["quantity"]*merged_data["price"]
```

```python
merged_data["order_date"] = pd.to_datetime(merged_data["order_date"])
```

```python
merged_data["year"] = merged_data["order_date"].dt.year
merged_data["month"] = merged_data["order_date"].dt.month
```

```python
# Task#01: A Complete Annual Report for every year (Total sale, Most selling category, most selling product)
orders2022 = merged_data[merged_data["order_date"].dt.year == 2022]
orders2023 = merged_data[merged_data["order_date"].dt.year == 2023]
orders2024 = merged_data[merged_data["order_date"].dt.year == 2024]

# Problem-1: Total Annual sale for each year
annual_sales_2022 = orders2022["total_sale"].sum()   #Total sales in 2022 year
print(f"Annual sales of year 2022: {annual_sales_2022:.2f}")

annual_sales_2023 = orders2023["total_sale"].sum()   #Total sales in 2023 year
print(f"Annual sales of year 2023: {annual_sales_2023:.2f}")

annual_sales_2024 = orders2024["total_sale"].sum()   #Total sales in 2024 year
print(f"Annual sales of year 2024: {annual_sales_2024:.2f}")
```

```python
# Problem-2: Top selling category in each year
def find_top_category(year):
    orders = merged_data[merged_data["order_date"].dt.year == year]

    accessories = orders[orders["category"] == "Accessories"]["quantity"].sum()
    clothing = orders[orders["category"] == "Clothing"]["quantity"].sum()
    electronics = orders[orders["category"] == "Electronics"]["quantity"].sum()

    top_quantity = max(accessories, clothing, electronics)
    top_category = (
        "Accessories" if top_quantity == accessories
        else "Clothing" if top_quantity == clothing
        else "Electronics")

    print(f"Top category in {year} is: {top_category} , Quantity sold: {top_quantity}")
```

```python
top_most_category_2022 = find_top_category(2022)  #Top most selling category in 2022 year
top_most_category_2023 = find_top_category(2023)  #Top most selling category in 2023 year
top_most_category_2024 = find_top_category(2024)  #Top most selling category in 2024 year
```

```python
# Problem-3: Top selling product name 
orders2022_sorted = orders2022.sort_values("quantity", ascending=False) 
top_product_2022 = orders2022_sorted.iloc[0]  #Top selling product details in 2022
print(f"Most selling product in 2022: {top_product_2022["product_name"]}, Quantity: {top_product_2022["quantity"]}")

orders2023_sorted = orders2023.sort_values("quantity", ascending=False)
top_product_2023 = orders2023_sorted.iloc[0]  #Top selling product details in 2023
print(f"Most selling product in 2023: {top_product_2023["product_name"]}, Quantity: {top_product_2023["quantity"]}")

orders2024_sorted = orders2024.sort_values("quantity", ascending=False)
top_product_2024 = orders2024_sorted.iloc[0]  #Top selling product details in 2024
print(f"Most selling product in 2024: {top_product_2024["product_name"]}, Quantity: {top_product_2024["quantity"]}")
```

```python
# Task#02: A Complete Monthly Report for every month (Total sale, Most selling category, most selling product)
def monthly_report(year, month):
    # Filter data for the given year and month
    monthly_data = merged_data[(merged_data["year"] == year) & (merged_data["month"] == month)]

    if monthly_data.empty:
        print(f"No data found for {month}/{year}")
        return

    total_sale = monthly_data["total_sale"].sum()

    most_selling_category = monthly_data["category"].value_counts().idxmax()

    top_product_row = monthly_data.sort_values("quantity", ascending=False).iloc[0]
    most_selling_product = top_product_row["product_name"]
    product_quantity = top_product_row["quantity"]

    print(f"Report for {month}/{year}")
    print(f"Total Sale: ${total_sale:.2f}")
    print(f"Most Selling Category: {most_selling_category}")
    print(f"Most Selling Product: {most_selling_product}")
    print(f"Quantity Sold: {product_quantity}")
    
for year in range(2022, 2025):  #From 2022 to 2024 Year
    print(f"=========================== Year:{year} ===========================")
    for month in range(1, 13):  #Months 1 to 12 
        monthly_report(year, month)
        print("================================")
```

```python
# Problem-3. Most re-visited customer
most_visited_customers = merged_data["customer_id"].value_counts()
top_revisited_customer_id = most_visited_customers.idxmax()
customer_name = merged_data[merged_data["customer_id"] == top_revisited_customer_id]["name"].iloc[0]
visits = most_visited_customers.max()

print(f"Most Revisited Customer ID: {top_revisited_customer_id}")
print(f"Customer Name: {customer_name} , Total visits: {visits}")
```

```python
# Problem-4: Most selling category in all 3-years
total_categories = merged_data["category"].value_counts().reset_index()
total_categories.columns = ["category", "count"]
top_category_name = total_categories["category"].iloc[0]
top_category_quantity = total_categories["count"].iloc[0]

print(f"Most Selling Category in 3 Years: {top_category_name}")
print(f"Total Quantity Sold: {top_category_quantity}")
```

```python
# Problem-5: List of all categories 
product_categories_list = products.category.unique()
print(f"All categories names: {product_categories_list}")
```

```python
# Problem-6: Customers who spent more than $2000 on our store
customer_sales = merged_data[["customer_id", "name", "price"]]
customer_sales = customer_sales.sort_values("customer_id")
unique_customers = customer_sales["customer_id"].unique()

high_spenders = []

for cid in unique_customers:
    data = customer_sales[customer_sales["customer_id"] == cid]    
    total_spent = data["price"].sum()
    if total_spent > 2000:
        name = data["name"].iloc[0]
        high_spenders.append([cid, name, total_spent])

high_spenders_df = pd.DataFrame(high_spenders, columns=["customer_id", "name", "total_spent"])
high_spenders_df.head()
```

```python
# Problem-7: Number of customers below 30, 50, 80 in every category
customer_category_age = merged_data[["customer_id", "category", "age"]].drop_duplicates()

# Count of customers in different age groups by category
age_below_30 = customer_category_age[customer_category_age["age"]< 30]
age_below_50 = customer_category_age[customer_category_age["age"]< 50]
age_below_80 = customer_category_age[customer_category_age["age"]< 80]

below_30_count = age_below_30["category"].value_counts() #Count of customers in each category for age < 30
below_50_count = age_below_50["category"].value_counts() #Count of customers in each category for age < 50
below_80_count = age_below_80["category"].value_counts() #Count of customers in each category for age < 80

age_groups_data = pd.DataFrame({
    "Below 30": below_30_count,
    "Below 50": below_50_count,
    "Below 80": below_80_count
})
age_groups_data
```
