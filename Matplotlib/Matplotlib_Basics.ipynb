```python
x = [1,2,3,4,5]
y1 = [i**2 for i in x]
y2 = [i**3 for i in x]

# Creating graph with two y-values
plt.title("Multi lines")
plt.xlabel("X axis")
plt.ylabel("Y axis")
plt.ylim(-200,200)
plt.xlim(-5,5)
plt.axhline(0) #Horizontal line
plt.axvline(1) #Vertical line
plt.plot(x,y1,label="y = x²",color="green")
plt.plot(x,y2,label="y = x³",color="red")
plt.legend()
plt.grid(True)
plt.savefig("multilinegraph.png")
plt.show()
```

### Customizing the plotting

```python
land_areas = [80,110,120,150,200]
prices = [30,50,60,90,100]

# plt.plot(land_areas,prices, color="green", linestyle="dashed")
# plt.plot(land_areas,prices, color="green", linestyle="dotted")
# plt.plot(land_areas,prices, color="green", linestyle="dotted",marker="o")  #marker heighlights our data values in graph (instersection line)
# plt.plot(land_areas,prices, color="green", linestyle="dotted",marker="o",markerfacecolor="red")  #markerfacecolor : to change the color our marker
plt.plot(land_areas,prices, color="green", linestyle="dotted",marker="o",markerfacecolor="red",markersize=10)  
plt.title("Price per square yard")
plt.xlabel("Square Yards")
plt.ylabel("Price in lacs")
plt.grid(True)
plt.show()
```

```python
grades = ["A","B","C","D"]
students_percentages = [30,50,20,10]

# plt.bar(grades,students_percentages,color="blue",facecolor="red")
plt.bar(grades,students_percentages,color=["blue","lightgreen","red","skyblue"],edgecolor=["black","red","purple","brown"],tick_label=["S_A","S_B","S_C","S_D"],width=0.5)  #tick_label: To change the horizontal labels 
plt.title("Students Grades")
plt.xlabel("grades")
plt.ylabel("percentages")
plt.show()
```

```python
plt.pie(students_percentages,labels = grades,autopct='%1.1f%%')
plt.show()
```
